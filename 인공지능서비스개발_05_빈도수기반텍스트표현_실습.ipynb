### 한글 문서 카운트 벡터 추출

```python
sample_corpus = [
    '자연어처리 강의를 시작하겠습니다.',
    '자연어처리는 재미있습니다.',
    '밥을 먹고 강의를 듣고 있습니다.',
    '이번 자연어처리 강의는 한국어 자연어처리입니다.' 
]
```

```python
from konlpy.tag import Okt
def my_tokenizer(text):
    my_tags =['Noun'] # , 'Verb', 'Adjective'
    my_stopwords = ['이번']
    return [ word for word, tag in Okt().pos(text) if tag in my_tags and word not in my_stopwords ]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Vectorizer 객체 생성
vectorizer = CountVectorizer(
    tokenizer = my_tokenizer
)

# 특징 집합 생성
vectorizer.fit(sample_corpus)
vectorizer.get_feature_names_out()

# 카운트 벡터 생성
# 특징 벡터 추출
dtm = vectorizer.transform(sample_corpus)
type(dtm) # scipy.sparse._csr.csr_matrix 희소행렬 
print(dtm.toarray())

# DTM 출력
```
